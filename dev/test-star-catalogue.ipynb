Conclusion:

- Star names are really a nightmare.
- We have all the name of stars approved by IAU in `IAU-CSN.json`. This include RA(J2000), Dec(J2000), mag. The file can be obtained from: https://raw.githubusercontent.com/mirandadam/iau-starnames/master/catalog_data/IAU-CSN.json. The source of this file is https://www.pas.rochester.edu/~emamajek/WGSN/IAU-CSN.txt. 
- We have a file with the proper motions of stars having HIP number. It comes from Stellarium (cross-id.dat).
- The data for the rest of the stars can be obtained from the HR catalogue.
- We can obtain also the data from the HIP catalogue: http://vizier.cds.unistra.fr/viz-bin/VizieR-3?-source=I/239/hip_main&-out.max=50&-out.form=HTML%20Table&-out.add=_r&-out.add=_RAJ,_DEJ&-sort=_r&-oc.form=sexa.
